```python
from pathlib import Path
import os

# Ensure backend package imports work when notebook is opened from project root
os.chdir(Path.cwd())
print('Working dir:', Path.cwd())
```

```python
from backend.core.graph import run_router

def ask(query: str):
    state = run_router(query)
    return {
        'query': query,
        'response': state['response'].content,
        'classification': state['classification'].model_dump(),
        'safety_flags': state['safety_flags'].model_dump(),
        'disclaimers': state['response'].disclaimers + state['response'].safety_notes,
        'request_id': state.get('request_id')
    }

print('Notebook helper ready: ask("your query")')
```

```python
# Example
result = ask('What are symptoms of diabetes?')
result
```
